```python
# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from deap import base, creator, tools, algorithms
# Определяем фиксированный набор параметров
n = 4 # количество параметров
Qmin = 1000 # минимальное требуемое качество продукции
# Написываем генератор данных
def generate_data():
    # Сгенерировать случайное количество операций от 10 до 20
    n_ops = np.random.randint(10, 21)
    # Создать пустой датафрейм для хранения данных
    data = pd.DataFrame()
    # Добавить столбец с номерами операций
    data['Operation'] = np.arange(1, n_ops + 1)
    # Для каждой операции сгенерировать случайное время выполнения от 10 до 60 минут
    data['Time'] = np.random.randint(10, 61, size=n_ops)
    # Для каждой операции сгенерировать случайные затраты ресурсов от 100 до 1000 рублей
    data['Cost'] = np.random.randint(100, 1001, size=n_ops)
    # Для каждой операции сгенерировать случайное качество продукции от 0 до 100 баллов
    data['Quality'] = np.random.randint(0, 101, size=n_ops)
    # Генерируем случайные коэффициенты для каждой операции
    coeffs = {
    'a': np.random.uniform(1, 10, n_ops),
    'b': np.random.uniform(0.1, 1, n_ops),
    'c': np.random.uniform(0.1, 1, n_ops),
    'd': np.random.uniform(0.1, 1, n_ops),
        
    'e': np.random.uniform(0.1, 1, n_ops),
    'f': np.random.uniform(10, 100, n_ops),
    'g': np.random.uniform(1, 10, n_ops),
    'h': np.random.uniform(1, 10, n_ops),
        
    'i': np.random.uniform(1, 10, n_ops),
    'j': np.random.uniform(1, 10, n_ops),
    'k': np.random.uniform(0.1, 1, n_ops),
    'l': np.random.uniform(0.1, 1, n_ops),
        
    'm': np.random.uniform(0.1, 1, n_ops),
    'n': np.random.uniform(0.1, 1, n_ops),
    'o': np.random.uniform(0.1, 1, n_ops)
    }
    # Создаем DataFrame на основе сгенерированных данных
    coeffs_df = pd.DataFrame(coeffs)
    # Сохраняем DataFrame в файл coeffs.csv
    coeffs_df.to_csv('data/coeffs.csv', index=False)
    # Вернуть данные
    return data
    
```
