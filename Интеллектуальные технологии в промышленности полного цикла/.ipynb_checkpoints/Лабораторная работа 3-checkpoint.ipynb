```python
# Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import linprog
# Определяем формат и размерность данных
columns_warehouse = ['X', 'Y', 'Vehicle', 'Supply'] # названия колонок для таблицы о складах
columns_client = ['X', 'Y', 'Demand'] # названия колонок для таблицы о клиентах
rows = 4 # количество строк (складов)
clients = 2 # количество клиентов
# Генерируем случайные или фиксированные значения для каждой колонки
```

```python
X = np.random.randint(0, 100, size=rows) # X координата склада
Y = np.random.randint(0, 100, size=rows) # Y координата склада
C = np.random.randint(0, 100, size=(clients, 2)) # координаты клиентов
Distance = np.zeros((rows, clients)) # расстояние от склада до клиента
for i in range(rows):
    for j in range(clients):
        Distance[i, j] = np.sqrt((X[i] - C[j, 0])**2 + (Y[i] - C[j, 1])**2) #вычисляем расстояние по теореме Пифагора
Vehicle = np.random.choice(['truck', 'van'], size=rows) # тип транспортного средства на складе
Supply = np.random.randint(10, 50, size=rows) # запас на складе
Demand = np.random.randint(10, 50, size=clients) # спрос у клиента
# Собираем данные о складах в одну таблицу
data_warehouse = pd.DataFrame(columns=columns_warehouse) # создаем пустую таблицу данных о складах
data_warehouse['X'] = X # добавляем колонку X
data_warehouse['Y'] = Y # добавляем колонку Y
data_warehouse['Vehicle'] = Vehicle # добавляем колонку Vehicle
data_warehouse['Supply'] = Supply # добавляем колонку Supply
data_warehouse.index = ['Warehouse' + str(i + 1) for i in range(rows)] #задаем индексы для строк
# Собираем данные о клиентах в одну таблицу
data_client = pd.DataFrame(columns=columns_client) # создаем пустую таблицу данных о клиентах
data_client['X'] = C[:, 0] # добавляем колонку X
data_client['Y'] = C[:, 1] # добавляем колонку Y
data_client['Demand'] = Demand # добавляем колонку Demand
data_client.index = ['Client' + str(j + 1) for j in range(clients)] # задаем индексы для строк
# Сохраняем сгенерированные данные в файлы формата csv
data_warehouse.to_csv('data/logistics_data_warehouse.csv', index=True) # сохраняемданные о складах в файл с указанием индексов
data_client.to_csv('data/logistics_data_client.csv', index=True) # сохраняем данные о клиентах в файл с указанием индексов
print('Данные успешно сгенерированы и сохранены в файлы logistics_data_warehouse.csv и logistics_data_client.csv') # выводим сообщение об успешном завершении
# Читаем данные из файлов csv
data_warehouse = pd.read_csv('data/logistics_data_warehouse.csv', index_col=0) #итаем данные о складах из файла
data_client = pd.read_csv('data/logistics_data_client.csv', index_col=0) # читаем данные о клиентах из файла
# Визуализируем данные на графике
plt.figure(figsize=(10, 10)) # создаем фигуру с заданным размером
plt.scatter(data_warehouse['X'], data_warehouse['Y'],
s=data_warehouse['Supply'], c='blue', label='Warehouses') # рисуем точки для складов с размером, пропорциональным запасу, и синим цветом
plt.scatter(data_client['X'], data_client['Y'], s=data_client['Demand'],c='red', label='Clients') # рисуем точки для клиентов с размером,пропорциональным спросу, и красным цветом
for i in range(len(data_warehouse)): # для каждого склада
    plt.annotate(data_warehouse.index[i], (data_warehouse['X'][i], data_warehouse['Y'][i])) # добавляем подпись с индексом склада
for j in range(len(data_client)): # для каждого клиента
    plt.annotate(data_client.index[j], (data_client['X'][j],
    data_client['Y'][j])) # добавляем подпись с индексом клиента
plt.xlabel('X coordinate') # подписываем ось X
plt.ylabel('Y coordinate') # подписываем ось Y
plt.legend() # добавляем легенду

plt.title('Logistics network') # добавляем заголовок
plt.show()
```

```python
# Формируем вектор c
c = [] # создаем пустой список для коэффициентов
for i in range(rows): # для каждого склада
    for j in range(clients): # для каждого клиента
        c.append(Distance[i, j] * (10 if Vehicle[i] == 'truck' else 15)) #добавляем коэффициент, равный произведению расстояния и стоимости перевозки
c = np.array(c) # преобразуем список в массив numpy
# Формируем матрицу A_eq и вектор b_eq
A_eq = [] # создаем пустой список для матрицы
b_eq = [] # создаем пустой список для вектора
for j in range(clients): # для каждого клиента
    row = [] # создаем пустой список для строки матрицы
    for i in range(rows): # для каждого склада
        for k in range(clients): # для каждого клиента
            if k == j: # если клиент совпадает с текущим
                row.append(1) # добавляем 1 в строку матрицы
            else: # если клиент не совпадает с текущим
                row.append(0) # добавляем 0 в строку матрицы
    A_eq.append(row) # добавляем строку в матрицу
    b_eq.append(Demand[j]) # добавляем спрос клиента в вектор
A_eq = np.array(A_eq) # преобразуем список в массив numpy
b_eq = np.array(b_eq) # преобразуем список в массив numpy
# Формируем матрицу A_ub и вектор b_ub
A_ub = [] # создаем пустой список для матрицы
b_ub = [] # создаем пустой список для вектора
for i in range(rows): # для каждого склада
    row = [] # создаем пустой список для строки матрицы
    for j in range(rows): # для каждого склада
        for k in range(clients): # для каждого клиента
            if j == i: # если склад совпадает с текущим
                row.append(1) # добавляем 1 в строку матрицы
            else: # если склад не совпадает с текущим
                row.append(0) # добавляем 0 в строку матрицы
    A_ub.append(row) # добавляем строку в матрицу
    b_ub.append(Supply[i]) # добавляем запас склада в вектор
A_ub = np.array(A_ub) # преобразуем список в массив numpy
b_ub = np.array(b_ub) # преобразуем список
```

```python
# Решаем линейную программу с помощью функции linprog
res = linprog(c, A_eq=A_eq, b_eq=b_eq, A_ub=A_ub, b_ub=b_ub, method='simplex') # вызываем функцию linprog с заданными параметрами и методом симплекс-метода
x = res.x # получаем оптимальное решение
cost = res.fun # получаем оптимальное значение целевой функции
# Выводим результаты на экран
print('Оптимальное решение:')
for i in range(rows): # для каждого склада
    for j in range(clients): # для каждого клиента
        print(f'От склада {data_warehouse.index[i]} к клиенту {data_client.index[j]} перевезти {x[i * clients + j]:.2f} единиц груза') #выводим количество груза, перевозимого от склада к клиенту
print(f'Минимальные затраты на перевозку: {cost:.2f} рублей') # выводим минимальные затраты на перевозку
```

```python
# Визуализируем оптимальные маршруты на графике
plt.figure(figsize=(10, 10)) # создаем фигуру с заданным размером
plt.scatter(data_warehouse['X'], data_warehouse['Y'],
s=data_warehouse['Supply'], c='blue', label='Warehouses') # рисуем точки для складов с размером, пропорциональным запасу, и синим цветом
plt.scatter(data_client['X'], data_client['Y'], s=data_client['Demand'], c='red', label='Clients') # рисуем точки для клиентов с размером, пропорциональным спросу, и красным цветом
for i in range(len(data_warehouse)): # для каждого склада
    plt.annotate(data_warehouse.index[i], (data_warehouse['X'][i], data_warehouse['Y'][i])) # добавляем подпись с индексом склада
for j in range(len(data_client)): # для каждого клиента
    plt.annotate(data_client.index[j], (data_client['X'][j],data_client['Y'][j])) # добавляем подпись с индексом клиента
for i in range(rows): # для каждого склада
    for j in range(clients): # для каждого клиента
        if x[i * clients + j] > 0: # если от склада к клиенту перевозится положительное количество груза
            plt.plot([data_warehouse['X'][i], data_client['X'][j]],[data_warehouse['Y'][i], data_client['Y'][j]], c='green', linestyle='--',label=f'{x[i * clients + j]:.2f} units from {data_warehouse.index[i]} to{data_client.index[j]}') # рисуем линию от склада к клиенту зеленым цветом ипунктирным стилем с подписью количества груза и индексов склада и клиента
plt.xlabel('X coordinate') # подписываем ось X
plt.ylabel('Y coordinate') # подписываем ось Y
plt.legend() # добавляем легенду
plt.title('Optimal routes') # добавляем заголовок
plt.show() # показываем график
```
