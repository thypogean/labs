```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image
import numpy as np
```

```python
image=cv2.imread('kit.jpg')
# cv2.imshow('Image', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
plt.imshow(image)
```

```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
plt.imshow(blurred_image, cmap=None)
```

```python
gray_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
ret,thresh_img = cv2.threshold(gray_image, 180, 255, cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(thresh_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
img_contours = np.uint8(np.zeros((gray_image.shape[0],gray_image.shape[1])))

plt.imshow(cv2.drawContours(img_contours, contours, -1, (255,255,255), 1))
```

```python
faces = cv2.imread('80_02.jpg')
gray_img = cv2.cvtColor(faces, cv2.COLOR_BGR2GRAY) 
classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces_rect = classifier.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=9) 
```

```python
for (x, y, w, h) in faces_rect: 
    cv2.rectangle(faces, (x, y), (x+w, y+h), (0, 255, 0), thickness=1) 
```

```python
plt.imshow(faces)
```
