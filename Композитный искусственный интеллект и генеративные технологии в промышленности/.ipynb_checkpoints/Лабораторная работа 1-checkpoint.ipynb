Напишите программу, которая запрашивает у пользователя число n и выводит на экран n-й член последовательности Фибоначчи

```python
fib1 = 1
fib2 = 1
 
n = input("Номер элемента ряда Фибоначчи: ")
n = int(n)
 
i = 0
while i < n - 2:
    fib_sum = fib1 + fib2
    fib1 = fib2
    fib2 = fib_sum
    i = i + 1
print("Значение этого элемента:", fib2)
```
