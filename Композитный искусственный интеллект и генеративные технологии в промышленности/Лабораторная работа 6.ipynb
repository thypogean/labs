```python
from torchvision import datasets, transforms
# Определение трансформеров для нормализации данных
transform = transforms.Compose([
transforms.ToTensor(),
transforms.Normalize((0.1307,), (0.3081,))
])
# Загрузка набора данных MNIST
train_dataset = datasets.MNIST(root='data', train=True, download=True,
transform=transform)
test_dataset = datasets.MNIST(root='data', train=False, download=True,
transform=transform)

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, 
                                           batch_size=16, 
                                           shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, 
                                           batch_size=16, 
                                           shuffle=False)
```

```python
import torch.nn as nn
import torch.nn.functional as F
class FNN(nn.Module):
    def __init__(self):
        super(FNN, self).__init__()
        self.layer1 = nn.Linear(784, 256)
        self.layer2 = nn.Linear(256, 128)
        self.layer3 = nn.Linear(128, 10)
    def forward(self, x):
        x = torch.flatten(x, 1)
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        x = self.layer3(x)
        return x
```

```python
model = FNN()
```

```python
import torch.optim as optim
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 4
for epoch in range(num_epochs):
    for x, (images, labels) in enumerate(train_loader): 
        imgs = images.reshape(-1, 28*28)
        output = model(imgs)
        losses = criterion(output, labels)
        optimizer.zero_grad()
        losses.backward()
        optimizer.step()
        
    if (x+1) % 100 == 0:
        print(f'Epochs [{epoch+1}/{numepchs}], Step[{x+1}/{nttlstps}], Losses: {losses.item():.4f}')
```
