```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('kit.jpg')
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
mask = cv2.inRange(hsv, (50, 30, 20),(230,230,253))
plt.imshow(image)
plt.show()
plt.imshow(mask)
```

```python
src_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
dst = cv2.Laplacian(src_gray, cv2.CV_16S, ksize=5)
plt.imshow(dst, cmap ='gray')
```

```python
import numpy as np

kernel = np.array(
    [[1,0,-1],
     [1,0,-1],
     [1,0,-1]]
)
img = cv2.filter2D(src=image, ddepth=cv2.CV_16S, kernel=kernel)
plt.imshow(img)
```

```python
rows, cols = image.shape[:2]
   
# generating vignette mask using Gaussian 
# resultant_kernels
X_resultant_kernel = cv2.getGaussianKernel(cols,200)
Y_resultant_kernel = cv2.getGaussianKernel(rows,200)
   
#generating resultant_kernel matrix 
resultant_kernel = Y_resultant_kernel * X_resultant_kernel.T
   
#creating mask and normalising by using np.linalg
# function
mask = 255 * resultant_kernel / np.linalg.norm(resultant_kernel)
output = np.copy(image)
   
# applying the mask to each channel in the input image
for i in range(3):
    output[:,:,i] = output[:,:,i] * mask
plt.imshow(output)
```
